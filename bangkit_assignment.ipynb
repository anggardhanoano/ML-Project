```
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```
url = 'https://github.com/haysacks/midi-generator/raw/master/jazz-midi.csv'
df = pd.read_csv(url, index_col = 0)
print(df.shape)
df.head()
```

```
# before cleaning dataset column Notes
print(df["Notes"])
```

# Data Preparation

```
# Cleaning and use just first 100 data from dataset column Notes
notes = df["Notes"][0:100].str.replace("'", "").str.replace(' ', '').str.replace('{', '').str.replace('}', '').str.strip('[]').str.split(',')
print(notes)
```

```
# Convert the multidimensional list into 1D list to elimininate duplicate value
notes = [j for i in list(notes) for j in i]
print(notes)
```

```
# sort and eliminate duplicate notes, then got all pitch names
pitchnames = sorted(set(i for i in notes))
print(pitchnames)
```

```
# categories for all pitchnames, mapped from pitchnames list
notes_categories = dict((note,number) for number, note in enumerate(pitchnames))
print(notes_categories)
```

```
# sequence for predict 1 note of notes sequence
sequence_length = 50
length = len(notes) - sequence_length

# Input and output for LSTM
x = []
y = []

for i in range(length):
    sequence_in = notes[i : i + sequence_length]
    sequence_out = notes[i + sequence_length]
    
    x.append([notes_categories[note] for note in sequence_in])
    y.append(notes_categories[sequence_out])


n_patterns = len(x)

n_vocab = len(set(notes))
```

```
# reshape x to become 1 dimension
x = np.reshape(x, (n_patterns, sequence_length, 1))
print(x[0])
```

```
# Normalise x into 0 - 1 value
x = x / float(n_vocab)
print(x[0])
```

```
# Encoding categorical Output + one hot encode categorical

# Before OneHotEncode and reshape
y = np.array(y)
hot_encode = OneHotEncoder(sparse=False)
print(y)
```

```
# After OnehotEncode and reshape
y = y.reshape(len(y), 1)
y = hot_encode.fit_transform(y)
print(y)
```

```
# preparation complete when data splitted to training dataset and test dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```
